```python
import numpy as np
import astropy.io.fits as pf
import pylab as mplot
%pylab inline
```

```python
metadata = np.genfromtxt('/project2/chihway/chihway/shearcat/code/measurement/tile_DR3_1_1.csv', dtype='str', delimiter=",")[1:]
# print(metadata)
```

```python
from astropy.table import Table, vstack
CAT = []
for i in range(100):
    tile = metadata[i][0][2:-1]
    print(tile)
    
    shear_cat = pf.open('/project2/chihway/data/decade/shearcat_v1/metacal_output_'+tile+'.fits')[1].data
    shear_id = np.load('/project2/chihway/data/decade/shearcat_v1/ids_match_'+tile+'.npz')
    gold_mask = np.load('/project2/chihway/data/decade/coaddcat_v1/gold_mask_'+tile+'.npz')
    gold_mask_full = gold_mask['maskSE']*gold_mask['maskIMA']*(gold_mask['maskSG']>=2)
    shear_id_masked_gold = shear_id['ids'][gold_mask_full]
    #     shear_id_masked_mcal = shear_cat['id'][(shear_cat['mcal_T_ratio_noshear']>0.5)*(shear_cat['mcal_s2n_noshear']>10)]
    mask_joint = np.in1d(shear_cat['id'], shear_id_masked_gold)
    #     print(len(shear_id_masked_mcal[mask_joint])*1.0/0.5/60/60)
    #[(shear_cat['mcal_T_ratio_noshear']>0.5)*(shear_cat['mcal_s2n_noshear']>10)]
    
    tb = Table.read('/project2/chihway/data/decade/shearcat_v1/metacal_output_'+tile+'.fits', format='fits')[mask_joint]
    print(len(tb))
    CAT.append(tb)
```

```python
new = vstack(CAT)
print(len(new))
print(len(new)*1.0/(90*0.5*60*60))
# print(new)
new.write('metacal_test.fits', overwrite=True)
```
