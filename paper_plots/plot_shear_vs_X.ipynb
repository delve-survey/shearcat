```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
abspath = False
readpath = 'results/shear_vs_X'

if abspath:
    readpath = '/home/zhuoqizhang/delve_shear/results/shear_vs_X'
```

```python
def linear(x, k, b): 
    return x*k+b

def residual(param, func, x, y, var): 
    y_pred = func(param, x)
    return np.sum((y_pred-y)**2/var)

def linear_fit(x, y, var): 
    param, cov = curve_fit(linear, x, y, sigma=var)
    return param, cov
```

```python
def read_result(fname):
    result = np.load(f'{readpath}/{fname}.npy')
    bincenter = result[0]
    e1 = result[1]
    e2 = result[2]
    e1_err = result[3]
    e2_err = result[4]
    return bincenter, e1, e2, e1_err, e2_err
```

### SNR

```python
bin_center, e1, e2, e1_err, e2_err = read_result('SNR')

param1, cov1 = linear_fit(bin_center, e1-np.mean(e1), e1_err)
param2, cov2 = linear_fit(bin_center, e2-np.mean(e2), e2_err)

m1 = param1[0]/np.sqrt(cov1[0,0])
m2 = param2[0]/np.sqrt(cov2[0,0])

plt.errorbar(bin_center,e1-np.mean(e1),yerr=e1_err,label=r'$\langle e_1\rangle$, $m1/\sigma=$'+f'{np.round(m1,2)}',fmt='b.', capsize=2)
plt.plot(bin_center, linear(bin_center,param1[0],param1[1]),color='blue')

plt.errorbar(bin_center*1.02,e2-np.mean(e2),yerr=e2_err,label=r'$\langle e_2\rangle$, $m1/\sigma=$'+f'{np.round(m2,2)}',fmt='r.', capsize=2)
plt.plot(bin_center, linear(bin_center,param2[0],param2[1]),color='red')

plt.xscale('log')
plt.xlabel('SNR')
plt.legend()
plt.grid()
plt.ylabel(r'$\langle e\rangle$')
```
