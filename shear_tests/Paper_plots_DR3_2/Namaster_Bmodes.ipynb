```python
for i in [-999, 1, 2, 3, 4]:
    DIR = '/scratch/midway3/dhayaa/SHEARTESTS_20240901/tomobin_%d/' % i
    plt.figure(figsize = (5,3))

    plt.axhline(0, color ='k')

    B     = np.load(DIR + '/Bmode.npy')[:]
    Noise = np.load(DIR + '/Bmode_Noise.npy')[:]
    ell   = B[-1]
    res = B[3] - np.mean(Noise[:, 3], axis = 0)
    # res[:] = 0
    Cov = np.cov(Noise[:, 3].T)

    Nmodes = len(res)
    NJK = len(Noise); print(f"NFILES :", NJK)
    hartlap  = ((NJK - Nmodes - 2) / (NJK -  1))
    dodelson = 1 / (1 + (Nmodes - 3) * (NJK - Nmodes - 2) / (NJK - Nmodes - 1) / (NJK - Nmodes - 4) )

    inv = np.linalg.inv(Cov) * hartlap * dodelson
    chi2 = res @ inv @ res

    print("----------FULL BMODE----------")
    print("P-val:", 1 - stats.chi2(df = len(res)).cdf(chi2))
    print("Chi2, Chi2/ndof:", chi2, chi2/len(res))
    print("--------------------------------\n\n")

    plt.scatter(np.arange(Nmodes) + 1 , res * 1e5, s=13, color = COLORS[0])
    plt.errorbar(np.arange(Nmodes) + 1 , res * 1e5, yerr = np.sqrt(np.diagonal(Cov))  * 1e5, fmt = 'none', color = COLORS[0], lw = 0, elinewidth = 1)

    ticks = np.arange(0, Nmodes, 5)
    plt.xticks(ticks, [int(ell[t]) for t in ticks])
    plt.ylabel(r'$X_{\!-}$  [$10^{-5}$]', size = 13, labelpad = 10)
    plt.xlabel(r'$\ell$', size = 13, labelpad = 10);
    plt.show()
```
