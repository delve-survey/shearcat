```python
import numpy as np
import pylab as mplot, os
%pylab inline

# import matplotlib
# matplotlib.rcParams['font.family'] = "sans-serif"
# matplotlib.rcParams['font.sans-serif'] = "Times New Roman"

mplot.rc('text', usetex=True)
mplot.rc('font', family='serif')

if os.environ['USER'] == 'dhayaa':
    os.environ['PATH'] = f"{os.environ['PATH']}:/software/texlive-2022-el8-x86_64/bin/x86_64-linux/"
```

```python
OUT = np.load('../../shear_catalog/response_s2n_size/grid_quantities_20240827.npy')
```

```python
ww = OUT['weight']
ww = ww.reshape(20,20).T
size_ratio_grid2 = np.unique(OUT['T_ratio'])
s2n_grid2 = np.unique(OUT['SNR'])
mplot.figure(figsize=(4,3.5))
mplot.imshow(ww, cmap = 'magma', origin='lower', aspect='auto')
inds = [0, 4, 8, 12, 16, 19]
mplot.xticks(inds, np.round(size_ratio_grid2[inds], 1), rotation=45, fontsize='10', horizontalalignment='right')
mplot.yticks(inds, s2n_grid2[inds].astype(int), rotation=45, fontsize='10', horizontalalignment='right')
mplot.title('Weights', fontsize=15)
mplot.xlabel('$T/T_{\\rm PSF}$', fontsize=15)
mplot.ylabel('SNR', fontsize=15)
mplot.colorbar()

mplot.savefig('weights.pdf', bbox_inches='tight')
```

```python
R = np.average([OUT['R11'], OUT['R22']], axis = 0)
R = R.reshape(20,20).T
size_ratio_grid2 = np.unique(OUT['T_ratio'])
s2n_grid2 = np.unique(OUT['SNR'])
mplot.figure(figsize=(4,3.5))
mplot.imshow(R, cmap = 'magma', origin='lower', aspect='auto')
inds = [0, 4, 8, 12, 16, 19]
mplot.xticks(inds, np.round(size_ratio_grid2[inds], 1), rotation=45, fontsize='10', horizontalalignment='right')
mplot.yticks(inds, s2n_grid2[inds].astype(int), rotation=45, fontsize='10', horizontalalignment='right')
mplot.title('$\\langle R \\rangle$', fontsize=15)
mplot.xlabel('$T/T_{\\rm PSF}$', fontsize=15)
mplot.ylabel('SNR', fontsize=15)
mplot.colorbar()

mplot.savefig('R.pdf', bbox_inches='tight')
```

```python
N = OUT['Ncounts']
N = N.reshape(20,20).T
size_ratio_grid2 = np.unique(OUT['T_ratio'])
s2n_grid2 = np.unique(OUT['SNR'])
mplot.figure(figsize=(4,3.5))
mplot.imshow(N, cmap = 'magma', origin='lower', aspect='auto')
inds = [0, 4, 8, 12, 16, 19]
mplot.xticks(inds, np.round(size_ratio_grid2[inds], 1), rotation=45, fontsize='10', horizontalalignment='right')
mplot.yticks(inds, s2n_grid2[inds].astype(int), rotation=45, fontsize='10', horizontalalignment='right')
mplot.title('Number counts', fontsize=15)
mplot.xlabel('$T/T_{\\rm PSF}$', fontsize=15)
mplot.ylabel('SNR', fontsize=15)
cbar = mplot.colorbar()

from matplotlib.ticker import ScalarFormatter
formatter = ScalarFormatter(useMathText=True)
formatter.set_powerlimits((-2, 2))  # Enforces scientific notation for large/small values
cbar.ax.yaxis.set_major_formatter(formatter)

mplot.savefig('counts.pdf', bbox_inches='tight')
```

```python
S = OUT['sigma_e']
S = S.reshape(20,20).T
size_ratio_grid2 = np.unique(OUT['T_ratio'])
s2n_grid2 = np.unique(OUT['SNR'])
mplot.figure(figsize=(4,3.5))
mplot.imshow(S, cmap = 'magma', origin='lower', aspect='auto')
inds = [0, 4, 8, 12, 16, 19]
mplot.xticks(inds, np.round(size_ratio_grid2[inds], 1), rotation=45, fontsize='10', horizontalalignment='right')
mplot.yticks(inds, s2n_grid2[inds].astype(int), rotation=45, fontsize='10', horizontalalignment='right')
mplot.title('$\sqrt{ \langle \sigma^2_{e_{1,2}} \\rangle}$', fontsize=14)
mplot.xlabel('$T/T_{\\rm PSF}$', fontsize=15)
mplot.ylabel('SNR', fontsize=15)
mplot.colorbar()

mplot.savefig('sigmae2.pdf', bbox_inches='tight')
```
